## DATA VISUALIZATION ON IRIS DATASET 
### IRIS IS A MULTIVARIATE DATASET, WITH 3 VALUES IN THE TARGET COLUMN (SPECIES) : 1. IRIS-SETOSA, 2. IRIS-VERSICOLOR, 3. IRIS-VIRGINICA

### Importing necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the dataset:

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
df = df.drop(columns = 'Id')
```

# 4. Scatter Plot
## A graphical representation used to show the relationship between two variables.

```python
plt.figure(figsize=(4,4))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', 
                hue='Species', style='Species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

# 5. Box Plot
## It provides a quick summary of a dataset by showing its median, quartiles, and range.

```python
plt.figure(figsize=(4,4))
sns.boxplot(data=df, x='Species', y='SepalLengthCm', palette='Set2')
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

# 6. Pair Plot
## A pair plot primarily helps in visualizing relationships and distributions of multiple variables in a dataset. Key points:
### Shows Pairwise Scatterplots: Displays scatterplots for each pair of numerical variables.
### Highlights Variable Relationships: Reveals correlations and patterns between features.
### Displays Distributions: Diagonal plots show individual variable distributions (e.g., histograms or KDEs).

```python
sns.pairplot(df, hue='Species', palette='husl', diag_kind='hist')
plt.show()
```
