## DATA VISUALIZATION ON A BINARY DATASET
###  The dataset consists of 3 columns Roll No., Grades and Result. For grades A-E the result is Pass (i.e, 1) and for grade F the result is Fail (i.e, 0)

### Importing necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading the dataset

```python
df = pd.read_csv("Result.csv")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

## 1. Bar Plot - for distribution of grades
### This plot shows the distribution of grades in the dataset.
### It helps visualize how many students received each grade.

```python
grades = df['Grade'].value_counts()
```

```python
plt.bar(grades.index, grades.values, color = 'darkblue')
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

## 2. 1) Pie Chart - for grade distribution
###  To show the percentage or proportion of students receiving each grade.

```python
plt.pie(grades, labels=grades.index, autopct='%1.1f%%')
plt.title('Grade Proportions')
plt.show()
```

## 2. 2) Pie Chart - for result distribution
### To visualize the proportion of students passing (Result = 1) versus failing (Result = 0).

```python
result_counts = df['Result'].value_counts()
plt.pie(result_counts.values, labels=['Pass', 'Fail'], autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Result Distribution')
plt.show()
```

## 3. Histogram - Result distribution
### A histogram provides a clear visual summary of categorical or numerical data.
### It displays the number of occurrences (count) for each category of the "Result" column (e.g., 1 = Pass and 0 = Fail).

```python
plt.figure(figsize=(4,4))
plt.hist(df['Result'], bins = 2, color = 'pink', edgecolor = 'black')
plt.title('Result Distribution')
plt.xlabel('Result (1 = Pass, 0 = Fail)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Fail', 'Pass'])
plt.show()
```
